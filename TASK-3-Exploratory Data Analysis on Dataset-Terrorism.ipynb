3) ANALYSING THE RELATIONSHIP BETWEEN COLUMNS

```python
# countplot between year v/s no of attacks
plt.figure(figsize=(10,5))
sns.countplot(x="iyear",data=data)
plt.xticks(rotation=90)
plt.title("NUMBER OF TERRORIST ATTACKS THAT TOOK PLACE BETWEEN 1970-2017")
plt.show()
```

HENCE MAXIMUM TERRORIST ATTACKS TOOK PLACE IN 2014

```python
# countplot between month v/s no of attacks
plt.figure(figsize=(10,5))
sns.countplot(x="imonth",data=data,width=0.5)
plt.title("NUMBER OF TERRORIST ATTACKS THAT TOOK PLACE BASED ON MONTH")
plt.show()
```

HENCE MAXIMUM TERRORIST ATTACKS TOOK PLACE IN THE MONTH 5(MAY)

```python
# countplot between day v/s no of attacks
plt.figure(figsize=(10,5))
sns.countplot(x="iday",data=data)
plt.title("NUMBER OF TERRORIST ATTACKS THAT TOOK PLACE BASED ON DAY OF THE MONTH")
plt.show()
```

HENCE MAXIMUM TERRORIST ATTACKS TOOK PLACE ON 15TH DAY OF THE MONTH

```python
#pie chart of regions and terrorist attacks
plt.figure(figsize=(5,30))
p = data['region_txt'].value_counts().tolist()
mylabels = data['region_txt'].value_counts().index
plt.pie(p,labels=mylabels,labeldistance=1.2)
plt.legend(bbox_to_anchor=(1.7,1.6),loc='upper right')
plt.show()
```

HENCE HIGGEST TERRORIST ATTACKS TOOK PLACE IN MIDDLE EAST AND NORTH AFRICAN REGION

```python
# barplot between country vs count of attacks
plt.figure(figsize=(10,8))
sns.barplot(x=data['country_txt'].value_counts().values[:40],y=data['country_txt'].value_counts()[:40].index,data=data,palette='YlOrRd_r')
plt.title('COUNTRIES THAT ARE EFFECTED BY TERRORIST ATTACKS')
plt.show()
```

FROM ABOVE BARPLOT WE CAN SAY THAT 'IRAQ' IS MOST EFFECTED BY TERRORIST ATTACKS FOLLOWED BY PAKISTAN

```python
plt.figure(figsize=(10,8))
sns.barplot(x=data['city'].value_counts().values[1:30],y=data['city'].value_counts()[1:30].index,data=data,palette='Purples_r')
plt.title('CITIES THAT ARE EFFECTED BY TERRORIST ATTACKS')
plt.show()
```

FROM ABOVE BARPLOT WE CAN SAY THAT 'BAGHDAD' CITY IN IRAQ IS MOST EFFECTED BY TERRORIST ATTACKS

```python
# no of terrorist attacks that were succeeded and not succeeded
plt.figure(figsize=(10,5))
sns.countplot(x='success',data=data,width=0.4)
plt.title('NO OF TERRORIST ATTACKS SUCCEEDED AND NOT SUCCEEDED')
plt.show()
```

FROM ABOVE PLOT WE CAN IDENTIFY THAT MOST OF THE ATTACKS (AROUND 160000) PLANNED WERE SUCEEDED

```python
# pie chart on type of weapons used for attack
plt.figure(figsize=(5,10))
p = data['attacktype1_txt'].value_counts().tolist()
mylabels = data['attacktype1_txt'].value_counts().index
plt.pie(p,labels=mylabels)
plt.legend(bbox_to_anchor=(1.9,1.3),loc='upper right')
plt.show()
```

FROM ABOVE PIE CHART WE CAN SAY THAT MOST OF THE TERRORIST ATTACKS WERE 'BOMBING/EXPLOSION'

```python
# bar graph of targeted people (for whom the attack was planned)
plt.figure(figsize=(10,5))
sns.countplot(x='targtype1_txt',data=data,width=0.6)
plt.title('TARGETED PEOPLE')
plt.xticks(rotation=90)
plt.show()
```

THESE ATTACKS MOSTLY TARGETED 'PRIVATE CITIZENS AND PROPERTY'

```python

new_data = data.groupby(['iyear'],as_index=False)['nkill'].sum()
new_data.head()
```

```python
# year v/x no.people killed
plt.figure(figsize=(10,8))
plt.xticks(rotation=90)
plt.title("NO OF PEOPLE KILLED IN A PARTICULAR YEAR")
sns.barplot(data=new_data,x='iyear',y='nkill')
plt.show()
```

As no.of attacks were maximum in 2014 no.of people killed were also maximum in 2014

```python
new_data1 = data.groupby(['iyear'],as_index=False)['nwound'].sum()
new_data1.head()
```

```python
plt.figure(figsize=(10,8))
plt.xticks(rotation=90)
plt.title("NO OF PEOPLE WOUNDED IN A PARTICULAR YEAR")
sns.barplot(data=new_data1,x='iyear',y='nwound')
plt.show()
```

Intrestingly No.of people wounded is maximum in 2015 but not in 2014

```python
new_data = data.groupby(['region_txt'],as_index=False)['nkill'].sum()
new_data
```

```python
plt.title("NUMBER OF PEOPLE KILLED BASED ON REGION")
sns.barplot(data=new_data,y='region_txt',x='nkill')
plt.show()
```

```python
new_data1 = data.groupby(['region_txt'],as_index=False)['nwound'].sum()
new_data1
```

```python
plt.title("NUMBER OF PEOPLE WOUNDED BASED ON REGION")
sns.barplot(data=new_data1,y='region_txt',x='nwound')
plt.show()
```

Maximum number of people killed and wounded from 'MIDDLE EAST & NORTH AFRICA' region

```python
new_data2 = data.groupby(['gname'],as_index=False)['nkill'].sum().sort_values(['nkill'],ascending=False)
```

```python
new_data2
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=new_data2['gname'].values[2:20],y=new_data2['nkill'].values[2:20],data=new_data2,palette='YlOrRd_r')
plt.xticks(rotation=90)
plt.title('GROUPS ATTACKED V/S NUMBER OF PEOPLE KILLED')
plt.ylabel("NO OF PEOPLE KILLED")
plt.xlabel("GROUP NAME")
plt.show()
```

MOST WERE KILLED BY TALIBAN ATTACKS
